```python
import pandas as pd

file_path = r"D:\EU School\Academics\3rd Semester\Business Intelligence\Final Assessment\Stock-Market-Project\data\raw\AAPL.csv"

aapl = pd.read_csv(file_path, skiprows=2)

aapl.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']

aapl.head()

```

```python
aapl['Date'] = pd.to_datetime(aapl['Date'])
aapl.set_index('Date', inplace=True)
aapl.sort_index(inplace=True)
aapl.dropna(inplace=True)
aapl.info()
```

```python
aapl['% Change'] = aapl['Close'].pct_change() * 100

aapl[['Close', '% Change']].head(10)
```

```python
aapl['SMA_30'] = aapl['Close'].rolling(window=30).mean()

aapl['SMA_200'] = aapl['Close'].rolling(window=200).mean()

aapl[['Close', 'SMA_30', 'SMA_200']].tail(10)
```

```python
delta = aapl['Close'].diff()

gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

window = 14

avg_gain = gain.rolling(window=window).mean()
avg_loss = loss.rolling(window=window).mean()

rs = avg_gain / avg_loss
aapl['RSI'] = 100 - (100 / (1 + rs))

aapl[['Close', 'RSI']].tail(10)
```

```python
aapl['Volume_MA_20'] = aapl['Volume'].rolling(window=20).mean()

aapl[['Volume', 'Volume_MA_20']].tail(10)
```

```python
aapl['Volatility_30'] = aapl['% Change'].rolling(window=30).std()

aapl[['% Change', 'Volatility_30']].tail(10)
```

```python
final_columns = ['Open', 'High', 'Low', 'Close', 'Volume',
                 '% Change', 'SMA_30', 'SMA_200', 'RSI', 'Volume_MA_20', 'Volatility_30']

aapl_final = aapl[final_columns]

aapl_final.reset_index(inplace=True)

output_path = r"D:\EU School\Academics\3rd Semester\Business Intelligence\Final Assessment\Stock-Market-Project\data\processed\AAPL_cleaned.csv"
aapl_final.to_csv(output_path, index=False)

print("Cleaned dataset saved to:", output_path)
```

# Analysis for MSFT

```python
import pandas as pd

msft_path = r"D:\EU School\Academics\3rd Semester\Business Intelligence\Final Assessment\Stock-Market-Project\data\raw\MSFT.csv"
msft = pd.read_csv(msft_path, skiprows=2)

msft.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']

msft['Date'] = pd.to_datetime(msft['Date'])
msft.set_index('Date', inplace=True)
msft.sort_index(inplace=True)
msft.dropna(inplace=True)

msft.head()
```

## % Daily Price Change

```python
msft['% Change'] = msft['Close'].pct_change() * 100
```

## 30-day & 200-day SMA

```python
msft['SMA_30'] = msft['Close'].rolling(window=30).mean()
msft['SMA_200'] = msft['Close'].rolling(window=200).mean()
```

## RSI (Relative Strength Index)

```python
delta = msft['Close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()

rs = avg_gain / avg_loss
msft['RSI'] = 100 - (100 / (1 + rs))
```

## Volume Moving Average

```python
msft['Volume_MA_20'] = msft['Volume'].rolling(window=20).mean()
```

## Volatility (30-day)

```python
msft['Volatility_30'] = msft['% Change'].rolling(window=30).std()
```

```python
msft.tail()
```

## Lets now save Cleaned MSFT Data

```python
msft_columns = ['Open', 'High', 'Low', 'Close', 'Volume',
                '% Change', 'SMA_30', 'SMA_200', 'RSI', 'Volume_MA_20', 'Volatility_30']

msft_final = msft[msft_columns]

msft_final.reset_index(inplace=True)

output_path = r"D:\EU School\Academics\3rd Semester\Business Intelligence\Final Assessment\Stock-Market-Project\data\processed\MSFT_cleaned.csv"
msft_final.to_csv(output_path, index=False)

print("MSFT cleaned dataset saved to:", output_path)
```
